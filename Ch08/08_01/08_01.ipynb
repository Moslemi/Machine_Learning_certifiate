#### (Multiple) linear regression on the enrollment data

```python
address = 'enrollment_forecast.csv'#'C:/Users/Lillian Pierson/Desktop/Exercise Files/Ch08/08_01/enrollment_forecast.csv'
enroll = pd.read_csv(address)
enroll.columns = ['year','roll','unem', 'hgrad', 'inc']
enroll.head()
```

```python
sb.pairplot(enroll)
```

```python
enroll_data = enroll.iloc[:,[2,3]].values
enroll_target = enroll.iloc[:,1].values
enroll_data_names = ['unem', 'hgrad']

X, y = scale(enroll_data), enroll_target  # we should always scale the data
```

### Checking for missing values

```python
missing_values = X==np.NAN # check wethere we have missing value or not by calling tis boolean function (True or False)
X[missing_values == True]
```

```python
LinReg = LinearRegression(normalize=True)

LinReg.fit(X,y)

print (LinReg.score(X,y))
```
